```python
class MyClass:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    @classmethod
    def from_string(cls, s):
        # Parse string and extract values for x and y
        x, y = map(int, s.split(','))
        # Return instance created using parsed values
        return cls(x, y)

    @classmethod
    def default(cls):
        # Return instance with default values
        return cls(0, 0)

# Usage examples
obj1 = MyClass(5, 10)  # Using regular constructor
obj2 = MyClass.from_string("3,7")  # Using from_string method
obj3 = MyClass.default()  # Using default method
print(obj3.x,obj3.y)

```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_details(self):
        return f"Name: {self.name}, Age: {self.age}"

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def get_details(self):
        return f"{super().get_details()}, Student ID: {self.student_id}"

class Professor(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def get_details(self):
        return f"{super().get_details()}, Employee ID: {self.employee_id}"

# Example usage:
student = Student("John Doe", 20, "S12345")
professor = Professor("Dr. Smith", 45, "P67890")

print(student.get_details())  # Output: Name: John Doe, Age: 20, Student ID: S12345
print(professor.get_details()) # Output: Name: Dr. Smith, Age: 45, Employee ID: P67890
```

```python
Encapsulation is an object-oriented programming (OOP) concept that binds together the data (attributes) and the code (methods) that manipulates the data, restricting direct access to some of an object’s components. This is done to prevent the accidental modification of data and to promote data integrity. Encapsulation allows you to hide the internal state of an object and only expose a controlled interface.
```

```python
Access Control: Encapsulation allows you to control access to the data by defining which attributes and methods are publicly accessible, and which are hidden from external access.
Data Hiding: Encapsulation enables the hiding of the internal implementation details of a class. This prevents external entities from depending on or interfering with the internal workings of the object.
```

```python
class Example:
    def __init__(self, value):
        self.public = value  # Public attribute
        self._protected = value  # Protected attribute
        self.__private = value  # Private attribute

    def get_private(self):
        return self.__private

# Example usage:
example = Example(42)
print(example.public)  # Output: 42
print(example._protected)  # Output: 42
print(example.get_private())  # Output: 42
# print(example.__private)  # This would raise an AttributeError
```

```python
class Person:
    def __init__(self, age):
        self.__age = age  # Private attribute

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            raise ValueError("Age must be positive")

# Example usage:
person = Person(30)
print(person.get_age())  # Output: 30
person.set_age(25)
print(person.get_age())  # Output: 25
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number  # Private attribute
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            raise ValueError("Invalid withdrawal amount")

    def get_balance(self):
        return self.__balance

# Example usage:
account = BankAccount("123456789", 1000)
account.deposit(500)
print(account.get_balance())  # Output: 1500
account.withdraw(200)
print(account.get_balance())  # Output: 1300
```

```python
Polymorphism in Python refers to the ability of different objects to be treated as instances of the same class through a common interface. In object-oriented programming, polymorphism allows methods to be defined in a base class that can be overridden by derived classes, thus enabling different behaviors for the same method depending on the object that is calling it.
```

```python
Compile-time Polymorphism: Also known as method overloading, it occurs when multiple methods with the same name but different parameters are defined. Python does not natively support compile-time polymorphism since it does not allow method overloading in the traditional sense. However, this can be emulated using default parameters or variable-length arguments.

Runtime Polymorphism: This occurs when a method in a base class is overridden in a derived class, allowing for different implementations. The method to be executed is determined at runtime, depending on the object’s type. Python fully supports runtime polymorphism through method overriding.
```

```python
class Shape:
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def calculate_area(self):
        return 3.14 * (self.radius ** 2)

class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def calculate_area(self):
        return self.side * self.side

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def calculate_area(self):
        return 0.5 * self.base * self.height

# Example usage:
shapes = [Circle(5), Square(4), Triangle(3, 6)]
for shape in shapes:
    print(f"The area is {shape.calculate_area()}")
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        return "Car engine starting"

# Example usage:
car = Car()
print(car.start())  # Output: Car engine starting
```

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

# Example usage:
shapes = [Circle(5), Rectangle(4, 6)]
for shape in shapes:
    print(f"The area is {shape.area()}")
```

```python
class Animal:
    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        original = super().speak()
        return original + " Woof!"

# Example usage:
dog = Dog()
print(dog.speak())  # Output: Some sound Woof!
```

```python
class Account:
    def withdraw(self, amount):
        pass

class SavingsAccount(Account):
    def withdraw(self, amount):
        return f"Withdrawing {amount} from Savings Account"

class CheckingAccount(Account):
    def withdraw(self, amount):
        return f"Withdrawing {amount} from Checking Account"

class CreditCardAccount(Account):
    def withdraw(self, amount):
        return f"Withdrawing {amount} from Credit Card Account"

# Example usage:
accounts = [SavingsAccount(), CheckingAccount(), CreditCardAccount()]
for account in accounts:
    print(account.withdraw(100))
```

```python
Composition in Python is a design principle where one class contains an instance of another class, meaning a complex object is built from simpler objects. This "has-a" relationship allows objects to be composed of other objects, enabling modular and reusable code. For example, a Car might be composed of Engine, Wheels, and Transmission objects, each responsible for different parts of the car's functionality.
```

```python
Inheritance: Inheritance establishes an "is-a" relationship. It allows a new class (child) to inherit the properties and methods of an existing class (parent). This is useful when there is a clear hierarchical relationship between classes. For instance, a Dog "is-a" Animal.

Composition: Composition establishes a "has-a" relationship, where one class is composed of one or more objects of other classes. This is useful when classes should collaborate but don't share a parent-child relationship. For instance, a Car "has-a" Engine.
```

```python
class Author:
    def __init__(self, name, birthdate):
        self.name = name
        self.birthdate = birthdate

class Book:
    def __init__(self, title, author, publication_date):
        self.title = title
        self.author = author  # Composition: Book "has-a" Author
        self.publication_date = publication_date

    def book_info(self):
        return (f"'{self.title}' by {self.author.name}, published on {self.publication_date}.")

# Example usage:
author = Author("George Orwell", "1903-06-25")
book = Book("1984", author, "1949-06-08")
print(book.book_info())
```

```python
Flexibility: Composition is more flexible than inheritance because it allows the combination of different objects to create complex behaviors without requiring a rigid class hierarchy.

Reusability: Objects composed of other objects can be reused in different contexts. Changing or extending functionality can be done by altering the composed objects rather than modifying the entire class.

Decoupling: Composition reduces the dependency between classes, making the system more modular and easier to maintain.
```

```python
class Room:
    def __init__(self, name, area):
        self.name = name
        self.area = area

class House:
    def __init__(self, address):
        self.address = address
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def total_area(self):
        return sum(room.area for room in self.rooms)

# Example usage:
living_room = Room("Living Room", 20)
kitchen = Room("Kitchen", 15)

house = House("123 Elm St")
house.add_room(living_room)
house.add_room(kitchen)

print(f"Total house area: {house.total_area()} square meters.")
```

```python
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

    def total_duration(self):
        return sum(song.duration for song in self.songs)

# Example usage:
song1 = Song("Song A", "Artist 1", 3.5)
song2 = Song("Song B", "Artist 2", 4.0)

playlist = Playlist("My Playlist")
playlist.add_song(song1)
playlist.add_song(song2)

print(f"Total playlist duration: {playlist.total_duration()} minutes.")
```

```python


class CPU:
    def __init__(self, model, speed):
        self.model = model
        self.speed = speed  # in GHz

class RAM:
    def __init__(self, size):
        self.size = size  # in GB

class Storage:
    def __init__(self, type, capacity):
        self.type = type  # SSD or HDD
        self.capacity = capacity  # in GB

class Computer:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu  # Computer "has-a" CPU
        self.ram = ram  # Computer "has-a" RAM
        self.storage = storage  # Computer "has-a" Storage

    def system_info(self):
        return (f"CPU: {self.cpu.model} ({self.cpu.speed} GHz), "
                f"RAM: {self.ram.size} GB, "
                f"Storage: {self.storage.capacity} GB {self.storage.type}")

# Example usage:
cpu = CPU("Intel i7", 3.6)
ram = RAM(16)
storage = Storage("SSD", 512)

computer = Computer(cpu, ram, storage)
print(computer.system_info())



```

```python
Delegation in composition is when an object delegates some of its responsibilities to another object. This simplifies the design of complex systems by breaking them down into smaller, more manageable parts, each handling a specific piece of functionality.

For instance, in a Car class, the Engine might be responsible for starting the car, and the Car object would delegate the start operation to the Engine object.
class Engine:
    def start(self):
        return "Engine started"

class Car:
    def __init__(self, engine):
        self.engine = engine  # Delegation: Car delegates starting to the Engine

    def start(self):
        return self.engine.start()

# Example usage:
engine = Engine()
car = Car(engine)
print(car.start())
```

```python
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

class Wheels:
    def __init__(self, size):
        self.size = size  # Size in inches

class Transmission:
    def __init__(self, type):
        self.type = type  # Manual or Automatic

class Car:
    def __init__(self, engine, wheels, transmission):
        self.engine = engine  # Car "has-a" Engine
        self.wheels = wheels  # Car "has-a" Wheels
        self.transmission = transmission  # Car "has-a" Transmission

    def car_info(self):
        return (f"Engine: {self.engine.horsepower} HP, "
                f"Wheels: {self.wheels.size} inches, "
                f"Transmission: {self.transmission.type}")

# Example usage:
engine = Engine(300)
wheels = Wheels(18)
transmission = Transmission("Automatic")

car = Car(engine, wheels, transmission)
print(car.car_info())
```
